Now, some more relevant plotting for my texts:

```python
# all taken from homework 1 assignment - just in case I need it:
from nltk import FreqDist, RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')

with open('poeWorks.txt', 'r', encoding='utf8') as file:
    string = file.read()
poeToken = tokenizer.tokenize(string)
poe = nltk.Text(poeToken)

with open('loveCraftWorks.txt', 'r', encoding='utf8') as file:
    string = file.read()
loveToken = tokenizer.tokenize(string)
love = nltk.Text(loveToken)

# new stuff begins here!
poeFrequent = ["upon", "one", "could", "would", "even"]
nltk.draw.dispersion_plot(poe, poeFrequent)
plt.show()

nltk.draw.dispersion_plot(love, poeFrequent)
plt.show()

#for myself: Order goes Poe and then Lovecraft
```

```python
#loveFrequent excludes any words previously included on poeFrequent.
loveFrequent = ["like", "us", "great", "must", "seemed"]
nltk.draw.dispersion_plot(poe, loveFrequent)
plt.show()

nltk.draw.dispersion_plot(love, loveFrequent)
plt.show()
```
